### 01 - 奇偶數辨別（Odd or even）
#### 當使用者輸入一個的數字，它能夠辨別奇偶，並輸出檢驗結果給使用者。

> 題目：輸入一數字判斷奇數或偶數<br>
輸入(Input): 一個數字 <br>
輸出(Output):奇數或偶數

* 進階: 可以限制輸入一個介於一定範圍（例如 1 到 1000）的數字

````python
while True:
    x = input("Insert a number between 1 and 1000: ")

    if x.isdigit() == 0:
        print("Error: You did not insert a valid number. Please try again.")
        
        continue

    a = int(x)

    if a > 1000:
        print("Error: The number is too large. Please try again.")
        
    elif a < 1:
        print("Error: The number is too small. Please try again.")
        
    else:
        if a % 2 == 0:
            print("It is an even number")
            
        else:
            print("It is an odd number")
            
        break 
````

### 03 -  猜數字（Guess the number）
#### 請使用者猜一個介於範圍（例如 1 到 100）之間的數字，
#### 若使用者猜錯，就詢問他們想繼續玩還是退出；若使用者猜對，就顯示祝賀訊息，
#### 並統計使用者的嘗試次數。如果使用者輸入的數字超出設定範圍，就顯示錯誤提示。

![](data\guess.png)


````python
import random

rn = random.randint(1, 100)
at = 0

while True:
    
    x = input('Please guess a number between 1 and 100: ')
    at += 1
    
    print ('Total attempts:',int(at),'\n')
    
    if x.isdigit() == 0:
        print("Error: You did not input a valid number. Please try again.")
        
        continue
        
    if int(x) < 1 or int(x) > 100:
        print("The number has to be between 1 and 100. Please try again.")
        
        continue
    
    a = int(x)
            
    if a > rn:
            print('Try again, the number is smaller than your guess.')
            
            y = input('Do you want to continue? Insert 1 to continue.')
            
            if y.isdigit() == 0:
                print('Thank you for playing.')
                break
                
            if int(y) == 1:
                continue
                
            else:
                print('Thank you for playing.')
                break
    
    
            
    elif a < rn:
            print('Try again, the number is larger than your guess.')
            
            y = input('Do you want to continue? Insert 1 to continue.')
            
            if y.isdigit() == 0:
                print('Thank you for playing.')
                break
                
            if int(y) == 1:
                continue
                
            else:
                print('Thank you for playing.')
                break
    
    
    if a == rn:
            print('Correct. Congratulations!')
            print ('It took you',int(at),'attempts to guess the correct number.')
            
            break

````

### 09 - 回文判斷（Is a palindrome）
#### 請使用者輸入單字，判斷它是否為回文，也就是該單字前後對稱，例如 madam，從前讀到後或是從後讀到前的順序都是 madam。

> 題目：<br>
輸入(Input): 雨 滋 春 樹 碧 連 天  天 連 碧 樹 春 滋 雨 <br>
輸出(Output):The text you have entered is a palindrome!<br>
輸入(Input): 資工訓練班 <br>
輸出(Output):The text you have entered is not a palindrome.


````python
def cvlowercase():
    rawinput = input('\nPlease enter a word, phrase, or a sentence: ')
    rawstring = rawinput.lower()
    rawlist = list(rawstring)
    return rawlist

r1 = cvlowercase()
print(r1)
````

````python
def nopuncs(fullcharlist): 
    puncslist = [" ", "-", ".", ",", ":", ";", "!", "?", "'", "\""] 
    for char in puncslist:
        if char in  fullcharlist:
            fullcharlist.remove(char)
            return nopuncs(fullcharlist)
    #去除相關標點符號
    return fullcharlist

input_string = input('Please input a sentence: ')
char_list = list(input_string)
print("Original list:", char_list)

r2 = nopuncs(char_list)
rstr = ''.join(r2)
print('List after removing punctuation:',r2,"\nString after removing punctuation:", rstr)
````

````python
def palindromecheck(straightlist):
    reversedlist = straightlist[::-1] 
    if reversedlist == straightlist: 
        return "The text you have entered is a palindrome!" 
    else: 
        return "The text you have entered is not a palindrome." 
````

````python
def main():
    olist = cvlowercase()
    olist = nopuncs(olist)
    pchecker = palindromecheck(olist)
    print(pchecker)

print('\nPalindrome checker')
main()
````
